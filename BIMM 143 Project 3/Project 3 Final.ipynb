## Part 1: Introduction

### Scientific Question: Do the active sites in the PSEN1 gene differ in Humans, model organisms, and animals who are known to get Alzheimer’s?

The PSEN1 gene makes the protein Presenilin 1. This protein is one subunit of γ-secretase,
which serves as a proteolytic enzyme (cuts proteins into peptides).The Presenilin 1
subunit acts as the proteolytic subcomponent of γ-secretase. This complex serves many proteolytic roles
but one of its most well known roles is in the processing of the Amyloid Precursor Protein (APP). The Presenilin 1
subunit cuts APP into soluble Amyloid Precursor Proteins as well versions of the Amyloid-Beta (β) peptide.
A key pathogenic indicator for Alzheimer's, one of the most widely prevalent neurodegenerative disorders in the world,
is the build up of Aβ proteins. Therefore, the PSEN1 gene and protein is seen as potential target area
for research and therapeutics due to its role in creating Aβ proteins from APP.

Two amino acid residues are key to the proteolytic activity of the PSEN1 protein as indicated by Wolfe et al (1999). Mutations of two aspartate
residues at Asp257 and Asp385 in the human PSEN1 proteinn showed a decrease is Aβ production and resulted in β protein fragments present. This highlights the
importance of these two residues and the amino acids(AA) around it to the proteolytic function of the protein. In this report, these
two sites will be looked at among model organisms and other animals known to be prone to Alzheimers.

The NCBI gene database (link: https://www.ncbi.nlm.nih.gov/gene/?term=psen1) was used to get all gene sequence and protein sequence data
for all species studied in this report. The NCBI gene database is filled with genomic data of millions of species and which can
quickly be accessed by keyword queries. Furthermore, the database also contains other tools besides genomic data search such as
the BLAST tool which finds local alignments between genes.

### Hypothesis: The PSEN1 gene is conserved between many animals, and is implicated in many cases of Alzheimers and therefore the active sites in the PSEN1 gene are the same in Humans, Model Organisms, and Dolphins
The PSEN1 gene is conserved between species. For example, a quick search in the NCBI gene database
for the PSEN1 gene resulted in genomic and protein data for Homo Sapiens along with 271 orthologs.
For years, we have known that zebrafish, mice and nonhuman primates are key model organisms that we can use to understand
different aspects of biology, from gene expressions to behaviour modeling. Due to this reason, I chose to include,
Zebrafish, Mice, Rheseus Monkeys, and Chimpanzees in my analysis of the PSEN1 gene. Furthermore,
in a study done by Moore et al (2017), the authors found novel evidence that presented dolphins
as a natural model for Alzheimer's disease due to their long lifespans. This study includes different
species of dolphins as well to see the similarities and differences in the PSEN1 genes between Humans and dolphins.

To answer my scientific question and test my hypothesis, I first had to obtain gene and protein sequences
for the different species I was comparing the human PSEN1 gene with. To do this, I went on the NCBI
gene database and downloaded the FASTA files for the gene sequences, and the protein sequences for the PSEN1 gene
of each animal. After the data was collected, each species PSEN1 gene sequence was aligned pairwise with
the human PSEN1 gene. Dot plots for each pairwise alignment was created to visualize the alignments. After the gene
sequences were aligned, the protein sequences for each PSEN1 gene were aligned using the
MUSCLE tool on: https://www.ebi.ac.uk/Tools/msa/muscle/. This multiple sequence alignment was
then used to find each active site in the PSEN1 protein that was mentioned above. These sites
were then plotted on a sequence logo to see the probability of the amino acid residues for each
species analyzed.

```python
'''
Function to create dot plots for two given sequences
codede adopted from: http://biopython.org/DIST/docs/tutorial/Tutorial.html#sec384
#parameters:
-seqA: first sequence to be plotted
-seqB: second sequence to be plotted
-window_size: the amount of nucleotides to match with the other sequence
-seqA_name: the name of the first sequence
-seqB_name: the name of the second sequence
#output:
a dotplot of the two sequences
'''
def dotplotcreator(seqA, seqB, window_size, seqA_name, seqB_name):
    window = window_size #for the window size for the dot plot

    #holds the two sequences that will be plotted together
    dict_one = {}
    dict_two = {}

    #mapping the window-sized sub-sequences to locations and adding them to the dictionary
    for (seq, section_dict) in [ (seqA, dict_one), (seqB, dict_two)]:
        for i in range(len(seq) - window):
            section = seq[i : i + window]
            #try and catch block
            try:
                section_dict[section].append(i)
            except KeyError:
                section_dict[section] = [i]

    #find any sub sequences that match the window size
    matches = set(dict_one).intersection(dict_two)
    print("%i unique matches" % len(matches))

    #Creates lists of x and y coordinates for the dot plot
    x = []
    y = []
    for section in matches:
        for i in dict_one[section]:
            for j in dict_two[section]:
                x.append(i)
                y.append(j)

    #plotting the graph
    pylab.cla()  # clear any prior graph
    pylab.gray()
    pylab.style.use("ggplot") # the style for the graph
    pylab.scatter(x, y) #plotting the graph itself

    #the scales of the dot plot
    pylab.xlim(0, len(seqA) - window)
    pylab.ylim(0, len(seqB) - window)

    #labeling the axis and titles
    pylab.xlabel("%s (length %i bp)" % (seqA_name, len(seqA)))
    pylab.ylabel("%s (length %i bp)" % (seqB_name, len(seqB)))
    pylab.title("Dot plot using window size %i\n(allowing no mis-matches)" % window)

    #showing the plot itself
    pylab.show()
```

### 4.2 Sequence Logos of the genes

In order to find the active sites that were highlighted by Wolfe et al (1999), I had to do a visual inspection
of the multiple sequence alignment below to find the particular residue that was mentioned in the article. The active site
was also double checked and confirmed via the Uniprot database, which is a freely accessible source of protein sequence
and functional information.

For the multiple sequence alignment, the MUSCLE program was used to align the protein sequences. I gathered
all the fasta files for the protein sequences and added them into one larger file. I then ran the
MUSCLE tool via: https://www.ebi.ac.uk/Tools/msa/muscle/, which gave me the multiple sequence
alignment.

In order to create a sequence logo, a matrix or dataframe must be created first with all the different
sequence positions and the number of nucleotides or amino acids present in that position. The sequence logo
then creates the logo out of this matrix, adjusting size via the number of the same amino acids present at a
given location.

```python
#read in the aligned file. I used the MUSCLE program online to align the 8 protein sequences
alignment = AlignIO.read(open("./PSEN1 proteins/psen1_protein_aligned.txt"), "fasta")
print("Alignment length %i" % alignment.get_alignment_length())

#visualizing the multiple sequence alignment
for record in alignment:
    print(record.seq + " " + record.description)
```

```python
#choosing the first active site present in Uniprot at AA 257
logo_alignment = alignment[:, 270:300]
```

```python
#creating a list of the AA in and around the active site for each genome
seq_list = []
for alignments in logo_alignment:
    seq_list.append(str(alignments.seq))
```

```python
#matrix of counts of a nucleotide at a given position
counts_mat = lm.alignment_to_matrix(seq_list)
```

```python
#sets the index so we can have an accurate representation of where in the alignment our sequence logo is
counts_mat.index = counts_mat.index + 270
```

```python
#Sequence Logo of PSEN1 Protein: AA250-AA280(Human Genome)
ax = lm.Logo(counts_mat, color_scheme="dmslogo_funcgroup")
ax.ax.set_ylabel("Number of Sequences")
ax.ax.set_xlabel("Position in Alignment")
ax.ax.set_title("Sequence Logo of First Active Site")
```

```python
#choosing the second active site present in Uniprot at AA 384 and 385
logo_alignment_2 = alignment[:, 400:430]
```

```python
#creating a list of the AA in and around the active site for each genome
seq_list_2 = []
for alignments in logo_alignment_2:
    seq_list_2.append(str(alignments.seq))
```

```python
#convert the strings into a dataframe with the number of residues at a given position
counts_mat_2 = lm.alignment_to_matrix(seq_list_2)

#gets us the proper range of the sequence logo within the alignment
counts_mat_2.index = counts_mat_2.index + 400
```

```python
#sequence logo of the second active site AA370-AA390(Human Genome)
ax = lm.Logo(counts_mat_2, color_scheme="dmslogo_funcgroup")
ax.ax.set_ylabel("Number of Sequences")
ax.ax.set_xlabel("Position in Alignment")
ax.ax.set_title("Sequence Logo of Second Active Site")
```

## Part 4: Results and Analysis

From the pairwise alignments, multiple sequence alignments, dot plots and sequence logos we can see that
even though the genome alignments are not perfect, meaning that the gene sequences differ between Humans
and other species, the active sites mentioned in Wolfe et al (1999). are well-preserved in model organisms.
Taking a look at the dotplots that were generated along with the alignment scores, it is evident that
Humans and Chimpanzees have PSEN1 genes that are highly similar. The same thing can be said for the Human and
Rheseus Monkey PSEN1 genes. This observation makes sense as all three species are all primates and share
a common ancestor closer to the present time than with the other species analyzed in this study. Surprisingly,
the PSEN1 gene for the Zebrafish, and Mice aligned terribly with the human PSEN1 gene. Furthermore, the sequence
alignments for the dolphin species was not as terrible as it was for the zebrafish and mice, which may
provide further evidence that Dolphins may be one of the better natural models to study Alzheimer's disease
with.

Looking at the sequence logos, the active sites to look at within the sequences are the Aspartate located at position 281 in
the first sequence logo and the aspartate located at position 411 in the second sequence logo.
These were the residues that were highlighted by Wolfe et al (1999) and the Uniprot database as being
vital to the function of the Presenilin 1 protein. Surprisingly, the amino acid sequence is well preserved in and around the active sites in all the
species that were analyzed in this study. This suggests that transcription, and post-transcriptional
modifications play a major role in the creation of the Presenilin 1 protein. Further analysis
of the transcript of PSEN1 needs to be done in order to truly determine where the different gene
sequences come to form a similar PSEN1 protein.

#### Conclusion
The active sites studied in this report do not differ in the PSEN1 gene
in Humans, model organisms, and animals who are known to get Alzheimer’s.
