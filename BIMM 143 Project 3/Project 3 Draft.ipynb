```python
from Bio import SeqIO
from Bio import AlignIO
from Bio import Align
import logomaker as lm

```

Load in data:

Data obtained from NCBI database

```python
#LOADING IN GENES PSEN1

#PSEN1 Humans
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_Human/data/PSEN1_human_gene.fna', "fasta"):
    psen1_human_gene = record.seq
#PSEN1 Chimps
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_Chimp/data/PSEN1_chimp_gene.fna', 'fasta'):
    psen1_chimp_gene = record.seq
#PSEN1 Zebrafish
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_zebrafish/data/PSEN1_zebrafish_gene.fna', 'fasta'):
    psen1_zebrafish_gene = record.seq
#PSEN1 Rheseus Monkey
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_Rheseus_Monkey/data/PSEN1_RMonkey_gene.fna', 'fasta'):
    psen1_RMonkey_gene = record.seq
#PSEN1 Mouse
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_mouse/data/PSEN1_mouse_gene.fna', 'fasta'):
    psen1_mouse_gene = record.seq
#PSEN1 Yangze Dolphin
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_Yangze_Dolphin/data/gene.fna', 'fasta'):
    psen1_YangzeD_gene = record.seq
#PSEN1 White-side-Dolphin
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_white-side_dolphin/data/PSEN1_white-side_dolphin.fna', 'fasta'):
    psen1_whitesideD_gene = record.seq
#PSEN1 bottlenose Dolphin
for record in SeqIO.parse('./PSEN1 genes/ncbi_dataset_PSEN1_bottlenose_dolphin/data/gene.fna', 'fasta'):
    psen1_bottlenoseD_gene = record.seq
```

```python
#Check the length of each PSEN1 gene:
print("PSEN1 human length: ", len(psen1_human_gene))
print("PSEN1 chimp length: ", len(psen1_chimp_gene))
print("PSEN1 Rheseus Monkey length: ", len(psen1_RMonkey_gene))
print("PSEN1 zebrafish length: ", len(psen1_zebrafish_gene))
print("PSEN1 Yangze Dolphin length: ", len(psen1_YangzeD_gene))
print("PSEN1 whiteside Dolphin length: ", len(psen1_whitesideD_gene))
print("PSEN1 bottleNose Dolphin length: ", len(psen1_bottlenoseD_gene))
print("PSEN1 mouse length: ", len((psen1_mouse_gene)))


```

```python
#To help with aligning the genes
aligner = Align.PairwiseAligner()
```

```python
#Aligning the genes
psen1_human_chimp_alignment = aligner.align(psen1_human_gene, psen1_chimp_gene)
psen1_human_RMonkey_alignment = aligner.align(psen1_human_gene, psen1_RMonkey_gene)
psen1_human_zebrafish_alignment = aligner.align(psen1_human_gene, psen1_zebrafish_gene)
psen1_human_mouse_alignment = aligner.align(psen1_human_gene, psen1_mouse_gene)
```

Sequence Logos of the genes

```python
#read in the aligned file. I used MAFFT program online to align the 5 genomes
alignment = AlignIO.read(open("PSEN1_aligned.txt"), "fasta")
print("Alignment length %i" % alignment.get_alignment_length())
for record in alignment:
    print(record.seq + " " + record.description)
```

```python
#choosing random 20 NT sequence alignment of the PSEN1 genes for the sequence logo.
#Just chose the first 20 NT that were present in the all 5 animal genomes
logo_alignment = alignment[:, 451:471]
```

```python
#creating a list of the 20 NT from each genome
seq_list = []
for alignments in logo_alignment:
    seq_list.append(str(alignments.seq))
```

```python
#matrix of counts of a nucleotide at a given position
counts_mat = lm.alignment_to_matrix(seq_list)
```

```python
#Sequence Logo of PSEN1 Gene NT: 451-471 (Human Genome)
lm.Logo(counts_mat)
```
